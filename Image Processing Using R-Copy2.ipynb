 <b><h3> Brightness,Contrast And Gamma Correction </h3> </b>
 <br>
  <u><h5><b>If we denote by x an image, α+βx^γ is an enhanced image where the parameter α controls the brightness, β the contrast and γ the γ-factor of the transformed image.</u></h5></b>
 

Case:1- Controlling Brightness

```r
img_b= img1+0.8
print(img_b)
display(img_b)
hist(img_b)
```

Case-2- Contrast

```r
img_c= img1*0.4
display(img_c)
```

Gamma Factor Correction

```r
img_g<- img1^3
display(img_g)
hist(img_g)
```
